
# Autoregressive process of oreder 2 and yule walker equation

```python
import numpy as np
import cmath
import scipy
from spectrum import *
import matplotlib.pyplot as plt                                          
from statsmodels.tsa.arima_process import ArmaProcess                      
from statsmodels.regression.linear_model import yule_walker    
```

# Generate a white gussian noise

```python
mean = 0                                                         
std = 1                                                         
num_samples = 1000                                               
white_noise = numpy.random.normal(mean, std, size=num_samples)   

plt.figure(figsize=[25, 5]);                                     
plt.title("White Noise ")                                    
plt.plot(white_noise)                                       
plt.show()                                                       
```

# AR2 Coefficients

```python
ar2 = np.array([1, -0.1, -0.8])                                            
ma = np.array([1])   
simulated_AR2_data = ArmaProcess(ar2, ma).generate_sample(nsample=1000)    

plt.figure(figsize=[25, 5]);                                               
plt.plot(simulated_AR2_data)                                                
plt.title("Simulated AR(2) Process")                                  
plt.show()                                                                  

```

```python

ar2 = np.array([1, 0.1, -0.8])                                            
ma = np.array([1])   
simulated_AR2_data = ArmaProcess(ar2, ma).generate_sample(nsample=1000)   

plt.figure(figsize=[25, 5]);                                               
plt.plot(simulated_AR2_data)                                               
plt.title("Simulated AR(2) Process 2.80")                                 
plt.show()                                                                  
```

# The Yule Walker Equations for the AR Coefficients

```python

a1 , a2 = -0.1 , 0.8      
p1 =(-a1 + (cmath.sqrt((a1**2)-4*a2)))/2
p2 =(-a1 - (cmath.sqrt((a1**2)-4*a2)))/2
print(f'p1 = {p1:.2f}' ,'      ', f'p2 = {p2:.2f}')

```

```python

a1 , a2 = 0.1 , -0.8      #2.80
p1 =(-a1 + (cmath.sqrt((a1**2)-4*a2)))/2
p2 =(-a1 - (cmath.sqrt((a1**2)-4*a2)))/2
print(f'p1 = {p1:.2f}' ,'      ', f'p2 = {p2:.2f}')
#2.80
```

```python

a1 , a2 = -0.975 , 0.95      #2.81
p1 =(-a1 + (cmath.sqrt((a1**2)-4*a2)))/2
p2 =(-a1 - (cmath.sqrt((a1**2)-4*a2)))/2
print(f'p1 = {p1:.2f}' ,'      ', f'p2 = {p2:.2f}')

#2.81
```
